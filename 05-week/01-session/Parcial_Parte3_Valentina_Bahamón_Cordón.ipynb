```python
import pandas as pd #leer el dataset
import numpy as np #calculos
import matplotlib.pyplot as plt #graficos
import seaborn as sns #graficos (caja y bigotes)
```

```python
df = pd.read_csv('notas_estudiantes_programa_academico.csv')
```

```python
#preparar los datos
df.head() #cinco filas
```

```python
df.isnull().sum() #datos faltantes
```

```python
#agrupar por materias
materias_grupo= df.groupby('Materia')
print('estudiantes - materias')

for materia, grupo in materias_grupo:
  numero_estudiantes = len(grupo) #cantidad
  print(numero_estudiantes, " - ", materia)
```

```python
#media, mediana y desviación estandar
estadisticas_materias = df.groupby('Materia')['Nota_Definitiva'].agg([
    ('Media', 'mean'),#media o promedio
    ('Mediana', 'median'), #mediana
    ('Desviación estandar', 'std'), #desviación estadar
    ('Minimo', 'min'), #nota minima
    ('Maxima', 'max'), #nota maxima
    ('Estudiantes', 'count') #n° estudiantes
])
estadisticas_materias
```

```python
plt.figure(figsize=(15,8)) #tamaño de grafico

sns.boxplot(
    x='Nota_Definitiva',
    y='Materia',
    data=df,
    palette='viridis',
    fliersize=8 #tamaño de los puntos atipicos

)

plt.title('Distribución de Notas por Materia', fontsize=16)
plt.grid(True)
plt.xticks(np.arange(0, 5.1, 0.5))
plt.show()
```
