```python
# Ejemplo base: Distribución de salarios por tipo de empleo

# Agrupamos el salario promedio y la mediana por tipo de empleo
avg_salary_by_type = df.groupby("employment_type")["salary_in_usd"].mean()
median_salary_by_type = df.groupby("employment_type")["salary_in_usd"].median()

print("Promedio de salario por tipo de empleo:")
print(avg_salary_by_type)
print("\nMediana de salario por tipo de empleo:")
print(median_salary_by_type)

# Graficamos el salario promedio
avg_salary_by_type.plot(kind="bar", color="skyblue", figsize=(6,4))
plt.title("Salario promedio por tipo de empleo")
plt.xlabel("Tipo de empleo")
plt.ylabel("Salario en USD (promedio)")
plt.show()

# Graficamos el salario mediano
median_salary_by_type.plot(kind="bar", color="orange", figsize=(6,4))
plt.title("Salario mediano por tipo de empleo")
plt.xlabel("Tipo de empleo")
plt.ylabel("Salario en USD (mediana)")
plt.show()

# Comparación con boxplot para ver la dispersión de salarios en cada tipo
df.boxplot(column="salary_in_usd", by="employment_type", figsize=(6,4))
plt.title("Distribución de salarios por tipo de empleo")
plt.suptitle("")
plt.xlabel("Tipo de empleo")
plt.ylabel("Salario en USD")
plt.show()
```

## Análisis de los resultados del ejercicio

- **Promedio vs. Mediana**:  
  El salario promedio puede verse influenciado por valores atípicos (muy altos o muy bajos), 
  mientras que la mediana refleja un valor más representativo de la mayoría de empleados.  
  Comparar ambas medidas permite identificar posibles distorsiones en los datos.

- **Gráficas de barras**:  
  - La primera muestra el **salario promedio** por tipo de empleo.  
  - La segunda muestra la **mediana de salario**, que en algunos casos puede diferir del promedio.  
  - Si la diferencia entre promedio y mediana es grande, puede indicar la presencia de outliers.

- **Boxplot**:  
  El diagrama de caja permite observar la **dispersión y distribución** de los salarios.  
  Aquí se puede analizar:  
    - La amplitud del rango intercuartílico (IQR).  
    - Si existen valores extremos (outliers).  
    - Comparar la concentración de salarios entre tipos de empleo (ej. FT, PT, CT, FL).  

En conclusión, este análisis permite ver no solo cuánto pagan en promedio los diferentes tipos de empleo, 
sino también cómo se distribuyen los salarios dentro de cada categoría, identificando diferencias y posibles anomalías.
