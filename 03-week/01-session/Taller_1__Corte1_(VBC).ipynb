1. **Definición de Ciencia de Datos:**

La ciencia de datos es una **disciplina interdisciplinaria** que combina métodos científicos, matemáticas, estadística, programación y experiencia en dominios específicos para **analizar grandes volúmenes de datos** con el objetivo de extraer información significativa que permita **identificar patrones**, responder preguntas como “qué pasó” o “qué pasará”, y tomar decisiones informadas en diversas áreas, especialmente en negocios y tecnología. Esta ciencia utiliza técnicas avanzadas para transformar datos crudos en conocimientos accionables que pueden mejorar estrategias, productos y procesos en tiempo real.

2. **Roles en la Ciencia de Datos**

Los roles comunes que suelen encontrarse son:

*Científico de Datos (Data Scientist):* Desarrolla la ingeniería de datos, modelado y análisis de negocio, con habilidades científicas y de comunicación.

*Ingeniero de Datos (Data Engineer):* Construye y mantiene la arquitectura de datos para asegurar el acceso y procesamiento eficiente.

*Analista de Datos (Data Analyst):* Se enfoca en la interpretación y análisis para mejorar la toma de decisiones, usando herramientas como SQL y Python.

*Ingeniero de Machine Learning:* Especializado en la creación y despliegue de modelos de aprendizaje automático.

3. **¿Qué son los dataset?**


Un dataset **(conjunto de datos)** es una colección organizada y estructurada de datos, generalmente presentada en forma de filas y columnas, que facilita su análisis y procesamiento para extraer información relevante. Estos conjuntos pueden contener diversos tipos de datos, desde números y texto hasta imágenes o grabaciones, y son esenciales para la ciencia de datos y el aprendizaje automático, ya que proporcionan la base para modelar, analizar y tomar decisiones informadas.

```python
# Cargar y leer un archivo .csv

import pandas as pd  #para manipular y analizar datos, facilitando la importación, limpieza y transformación

df = pd.read_csv('inventario_tecnologia.csv')  #para leer el archivo .csv
df.head(8)  #para mostrar las primeras n filas de un conjunto de datos
```

5. Generar una gráfica con el anterior `df`


```python
#ESTA GRAFICA LA HICE YO, GUIANDOME DE LO QUE HIZO EL PROFE
#Grafica de Barras para saber el stock de cada categeoria

import matplotlib.pyplot as plt #esta es una libreria especifica para crear gráficos

plt.figure(figsize=(10, 6)) #crea un lienzo de un tamaño especifico (ancho, alto)
stock_por_categoria = df.groupby('Categoria')['Stock'].sum().sort_values(ascending=False) #agrupa los datos en categoria
 #(es decir si hay varios producto de una misma categoria entonces se agrupan de acuerdo a la categoria)
 #y luego se suma el stock que tenga cada categoria y se ordena de forma descendente
stock_por_categoria.plot(kind='bar', color='skyblue') #se hace el tipo de grafica y el color de la grafica de los datos escogidos
plt.title('Stock Total por Categoría de Producto')
plt.xlabel('Categoría')
plt.ylabel('Cantidad en Stock')
plt.xticks(rotation=45) #esto es toque coqueto para que las etiquetas de X sea vean diagonale y evitar que se sobrepongan
plt.tight_layout() #tambien es un toque coqueto para que en el grafico no se vean superposiciones
#Estas dos funciones ultimas no son tan necesarias, pero hacen ver limpio el grafico
plt.show()
```

```python
#ESTA GRAFICA FUE AYUDA CON IA DE GEMINI QUE ME DA EL COLAB
# Generar un grafico de dispersion con el df

plt.figure(figsize=(10, 6)) # es para crear un lienzo de un tamaño especifico

# Crear el grafico de dispersion usando las columnas 'Precio_USD' y 'Stock' del df
plt.scatter(df['Precio_USD'], df['Stock'])

plt.title('Precio vs. Stock de Productos') # agregar titulo al grafico
plt.xlabel('Precio (USD)') # agregar etiqueta al eje x
plt.ylabel('Cantidad en Stock') # agregar etiqueta al eje y
plt.grid(True) # agregar una cuadricula para facilitar la lectura
plt.show() # mostrar el grafico
```
